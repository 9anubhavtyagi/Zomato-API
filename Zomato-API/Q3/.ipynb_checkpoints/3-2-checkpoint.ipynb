# Plot the histogram of aggregate rating of restaurant( drop the unrated restaurant).

### Module Imports

```python
# import libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

### Load Indian Restaurants data

```python
# fetch data
df = pd.read_csv('zomato.csv')

# extract indian restaurant's data
df = df[df['Country Code'] == 1]

# df
```

### drop rows containing nan values of aggregate ratings

```python
# drop nan values of aggregate ratings
df.dropna(subset=['Aggregate rating'], inplace=True)

# df
```

# Plot histogram

## I have plot 3 variations of histogram

## 1.) for ratings as [0.0, 0.1, 0.2, ....... , 4.9, 5.0]

```python
x_axis_points = np.arange(51)/10

plt.hist(df['Aggregate rating'], edgecolor='black',bins=x_axis_points)
plt.xticks(x_axis_points)



plt.title('Restaurant Rating Histogram')
plt.ylabel('Number of Restaurants')
plt.xlabel('Ratings')

plt.show()
```

## 2.) for ratings  grouped as [0, 0.5, 1, 1.5, 2, 2.5, 3, 3.5, 4, 4.5, 5]

```python
x_axis_points = [0, 0.5, 1, 1.5, 2, 2.5, 3, 3.5, 4, 4.5, 5]

plt.hist(df['Aggregate rating'], edgecolor='black', bins=x_axis_points)
plt.xticks(x_axis_points)

plt.title('Restaurant Rating Histogram\n Variation: bins as [0, 0.5, 1, 1.5, 2, 2.5, 3, 3.5, 4, 4.5, 5]')
plt.ylabel('Number of Restaurants')
plt.xlabel('Ratings')

plt.show()
```

## 3.) rating bins automatically decided by python i.e., bins = "auto"

```python
x_axis_points = np.arange(6)

plt.hist(df['Aggregate rating'], bins = "auto", edgecolor = "black")
plt.xticks(x_axis_points)

plt.title('Restaurant Rating Histogram\n Variation: Auto bin size')
plt.ylabel('Number of Restaurants')
plt.xlabel('Ratings')


plt.show()
```
