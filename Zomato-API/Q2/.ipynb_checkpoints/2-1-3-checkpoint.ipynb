# 3.) Rating vs Average Cost of Restaurant

#### It is not mentioned that in question that we need to do analysis based on average cost of two person, or one person.
#### So I am doing analysis based on average cost of two person (already present in zomato.csv).

## Approach: Find average cost with respect to each rating then plot the bar graph to see result

### Module Imports

```python
# importing libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

### Load Indian restaurants data

```python
# load data
df = pd.read_csv('zomato.csv')

# extract indian restaurant's data
df = df[df['Country Code'] == 1]

# df
```

### drop nan rows for columns Aggregate rating, Votes

```python
# drop nan values for Aggregate rating, Votes
new_df = df.dropna(subset=['Aggregate rating', 'Average Cost for two'])

# df
```

### Find average cost rating wise

```python
# sum of cost, rating wise
rating_wise_total_cost = new_df.groupby('Aggregate rating')['Average Cost for two'].sum()

# rating count in data
rating_wise_count = new_df.groupby('Aggregate rating')['Aggregate rating'].count()

# average cost, rating wise
rating_wise_avg_votes = rating_wise_total_cost/rating_wise_count

# rating_wise_avg_votes
```

### Separate ratings and average cost

```python
# ratings
ratings = rating_wise_avg_votes.index.astype(str)

# average votes
avg_cost = rating_wise_avg_votes.values
```

## Plot the graphs

```python
# plotting bar graph
plt.bar(ratings, avg_cost, width=0.8)
plt.xticks(rotation = 90)

plt.xlabel('Ratings')
plt.ylabel('Average Votes')
plt.title('Bar Graph - Ratings vs Average Count')

plt.show()
```

```python
# plotting line graph
plt.plot(ratings, avg_cost)
plt.xticks(rotation = 90)

plt.xlabel('Ratings')
plt.ylabel('Average Votes')
plt.title('Line Graph - Ratings vs Average Count')

plt.show()
```

# Result
- For rating less than or around 2: Unable to predict trend.
- For rating between 2 and 3: Restaurant's cost generally decreased with increase in rating.
- For rating greater than 3: Restaurant's cost generally increased with increase in rating.
- It can be concluded that the group of restaurants having rating greater than '3.9' are expensive but serves good quality of food service.
