# Write a short detail analysis of how the rating is affected by restaurant due following features: Plot a suitable graph to explain your inference.

# 1.) Rating vs Number of Votes given Restaurant

## Approach: Find average votes with respect to each rating then plot the "bar graph and line graph" to see result

### Importing libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

### Load data related to indian restaurants

```python
# load data
df = pd.read_csv('zomato.csv')

# extract indian restaurant's data
df = df[df['Country Code'] == 1]

# df
```

### drop rows containing nan values in Votes, and Rating Column

```python
# drop nan values for Aggregate rating, Votes
new_df = df.dropna(subset=['Aggregate rating', 'Votes'])

# df
```

### Calculate average votes (rating wise)

```python
# sum of votes rating wise
rating_wise_votes_sum = new_df.groupby('Aggregate rating')['Votes'].sum()

# rating count in data
rating_wise_count = new_df.groupby('Aggregate rating')['Aggregate rating'].count()

# average votes rating wise
rating_wise_avg_votes = rating_wise_votes_sum/rating_wise_count

# rating_wise_avg_votes
```

### Separate ratings and average votes

```python
# ratings
ratings = rating_wise_avg_votes.index.astype(str)

# average votes
avg_votes = rating_wise_avg_votes.values.astype(int)
```

## Plot the graphs

```python
# plotting bar graph
plt.bar(ratings, avg_votes, width=0.8)
plt.xticks(rotation = 90)

plt.xlabel('Ratings')
plt.ylabel('Average Votes')
plt.title('Bar Graph - Ratings vs Average Votes')

plt.show()
```

```python
# plotting line graph
plt.plot(ratings, avg_votes)
plt.xticks(rotation = 90)

plt.xlabel('Ratings')
plt.ylabel('Average Votes')
plt.title('Line Graph - Ratings vs Average Votes')

plt.show()
```

# Result
- For rating less than 3: Number of votes decrease slightly or almost remain same.
- For rating greater than 3: Number of votes increase significantly, with increase in ratings
- It can be concluded that the group of restaurants having rating greater than '4' receive significantly more number of votes
